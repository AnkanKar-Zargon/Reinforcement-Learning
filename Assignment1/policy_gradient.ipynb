```python
from environment import GridWorld
from agent import Agent
import matplotlib.pyplot as plt
from matplotlib import axes
import seaborn as sns

env=GridWorld()
```

```python
sns.set_style("darkgrid")
```

##### For sigma = 0.1

```python
agent = Agent(env, 0.1)
mean_and_standard_deviation_per_trial, mean_and_standard_deviation_per_episode = agent.hill_search()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(30, 10),  facecolor= "#C1CDCD")
plt.plot(mean_and_standard_deviation_per_trial[0], label='Mean', color = 'darkblue')
plt.plot(mean_and_standard_deviation_per_trial[1], label='Standard Deviation', color = 'red')
plt.xlabel('Trials')
plt.ylabel('Mean and Standard Deviation')
plt.title("For Sigma = 0.1")
plt.grid(color = 'black')
plt.legend()
plt.show()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(30, 10),  facecolor= "#C1CDCD")
plt.plot(mean_and_standard_deviation_per_episode[0], label='Mean', color = 'darkblue')
plt.plot(mean_and_standard_deviation_per_episode[1], label='Standard Deviation', color = 'red')
plt.xticks(range(0, len(mean_and_standard_deviation_per_episode[0])))
plt.xlabel('Episodes')
plt.ylabel('Mean and Standard Deviation')
plt.title('Per Episode for Sigma = 0.1')
plt.grid(color = 'black')
plt.legend()
plt.show()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(30, 10),  facecolor= "#C1CDCD")
plt.errorbar(range(0, len(mean_and_standard_deviation_per_trial[0])), mean_and_standard_deviation_per_trial[0], yerr=mean_and_standard_deviation_per_trial[1], fmt='.', color = 'red')
plt.xlabel('Trials')
plt.ylabel('Mean')
plt.title('Mean and Standard Deviation Per Trial for Sigma = 0.1')
plt.legend()
plt.show()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(30, 10),  facecolor= "#C1CDCD")
plt.errorbar(range(0, len(mean_and_standard_deviation_per_episode[0])), mean_and_standard_deviation_per_episode[0], yerr=mean_and_standard_deviation_per_episode[1], fmt='o', color = 'red')
plt.xlabel('Episodes')
plt.ylabel('Mean')
plt.title('Mean and Standard Deviation Per Episode for Sigma = 0.1')
plt.legend()
plt.show()
```
