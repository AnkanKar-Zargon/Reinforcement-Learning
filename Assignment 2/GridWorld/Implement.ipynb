```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import axes
import seaborn as sns

from environment import GridWorldEnv
from reinforce import REINFORCE, BaselineREINFORCE
```

```python
# Initialize the GridWorld environment
env = GridWorldEnv()

# REINFORCE Algorithm
reinforce_agent = REINFORCE(env)
reinforce_rewards = reinforce_agent.train(2000)


sns.set_style("darkgrid")
plt.figure(figsize=(30, 10),  facecolor= "#C1CDCD")
plt.plot(reinforce_rewards, label='REINFORCE Rewards', color = 'darkblue')
plt.xlabel('Episodes')
plt.ylabel('Total Reward')
plt.title('Learning Curve for REINFORCE Algorithm on GridWorld')
plt.grid(color = 'black')
plt.legend()
plt.show()

# Baseline REINFORCE Algorithm
baseline_reinforce_agent = BaselineREINFORCE(env)
baseline_reinforce_rewards = baseline_reinforce_agent.train(2000)

sns.set_style("darkgrid")
plt.figure(figsize=(30, 10),  facecolor= "#C1CDCD")
plt.plot(baseline_reinforce_rewards, label='Baseline-REINFORCE Rewards', color = 'darkblue')
plt.xlabel('Episodes')
plt.ylabel('Total Reward')
plt.title('Learning Curve for Baseline-REINFORCE Algorithm on GridWorld')
plt.grid(color = 'black')
plt.legend()
plt.show()
```
